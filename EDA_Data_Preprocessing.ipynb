If the distributions of the features are highly skewed, we cannot directly apply the common outlier detection methods such as the quantiles, boxplots or the IQR. If we do that, we'd be introducing bias to our dataset. Hence we first look at the individual distributions of all the features of the training set.

We try to convert all the features to have a Normal distribution, by applying the **YeoJohnson transformation**. (Since many observations are 0, BoxCox transformation would fail.)

```python
from scipy.stats import yeojohnson
```

```python
# Applying the YeoJohnson transformation to all the features in the train set and storing them in a new dataframe
df = pd.DataFrame()
for col in train.drop('Y', 1).columns:
    df[col] = yeojohnson(train[col])[0]  # automatically finds out the optimal value of lambda
    
```

```python
# We now check the distributions of the features again
df.hist(bins=30, figsize=(20, 10),layout=(10, 6))
plt.show()
```

```python
# We add the target variable to this dataset
df['Y'] = train['Y']
```

We see that even after the log-transformation, the distributions of most of the features seem to be heavily skewed. We also check the class-specific distributions of the features. We see that some of the features behave differently with both the classes. We also create a correlation matrix to check if the log-transformed features have any correlation. If the non-diagonal elements are all near 0, we can conclude that the features are mostly non-correlated.

# Feature Selection
We now check for multicollinearity using the Variance Inflation Factor. We drop the features with a VIF > 5 one by one. Also, since we would require an intercept in the model, we shall also add a constant column of 1's in the data.

```python
import statsmodels.formula.api as smf
# Calculating the VIF
def vif_cal(input_data):   
    x_vars=input_data
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]] 
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=smf.ols(formula="y~x", data=x_vars).fit().rsquared  
        vif=round(1/(1-rsq),2)
        print (xvar_names[i], " VIF = " , vif)
```

```python
import statsmodels.api as sm
features = list(df.drop(['Y'],1).columns)
X = df[features]
X = sm.add_constant(X)  # since we need an intercept
```

```python
vif_cal(X)
```

We observe that the VIF values of the features X32, X34, X55, X56 and X57 are higher than 5. We drop these features one by one and check the VIF of all the remaining features at each stage. Since the VIF value of X32 was the highest, we remove that first.

```python
vif_cal(X.drop(['X32'],1))
```

We notice that the VIF value of X34 has reduced, but that of X55, X56 and X57 are still greater than 5. We now drop X56 and recheck the VIFs of the remaining features.

```python
vif_cal(X.drop(['X32', 'X56'],1))
```

We observe that the VIF of X57 is still greater than 5, and hence we drop that feature as well.

```python
vif_cal(X.drop(['X32','X56','X57'],1))
```

After dropping the three features X32, X56 and X57, the VIFs of all the remaining features are lesser than 5, suggesting that the issue of multicollinearity has been dealt with. 

# Benchmark Model using Logistic Regression
Now, oftentimes, it so happens that not all of the given features significantly affect the outcome. Hence, these features just add to the complexity of the model, without adding any value. These features can be found out using their p-values. These features can also be dropped. Hence we create a benchmark model using Logistic regression with all the features excluding the ones removed in the previous step, and check the p-values of all these features.

We also split the training set into a train set and a validation test in the ratio of 4:1 as suggested in the problem statement.

```python
from sklearn.model_selection import train_test_split, cross_val_predict
coll_cols = ['X32', 'X56', 'X57'] # The columns to be dropped due to multicollinearity 
X_new = X.drop(coll_cols,1) # dropping the aforementioned columns 
Y = df['Y']
Y = Y.astype('int')
X_train, X_test, Y_train, Y_test = train_test_split(X_new, Y, test_size=0.20, random_state=42) # test size = 0.20 so that we get the ratio to be 4:1

# we Convert the validation set into a dataframe and store it as a csv

val_set = X_test.copy()
val_set['y_actual'] = Y_test
val_set.to_csv("Validation_set.csv", index = False)
```

We now fit a benchmark Logistic Regression model using the Logit function from statsmodels

```python
import statsmodels.api as sm

model = sm.Logit(Y_train , X_train)
fitted = model.fit()
fitted.summary()
```

We get a warning saying that the maximum number of iterations have been exceeded. We also get a note saying that there might be a complete quasi-separation and a fraction of observations can be perfectly predicted. This basically means that the target variable is able to separate one or more predictor variables completely, ie, we either have a case where the observations with $Y = 0 $ have some $X_i \le \alpha$ and $Y = 1$ have $X_i \ge \beta$, or vice versa. This could also be due to a combination of predictor variables. In such cases, the maximum likelihood estimate doesn't exist for those variables/features.

Though this method doesn't explicitly inform us about the variables causing the quasi-separation, we can get an idea by looking at the coeff , std error and the Wald confidence intervals.

For example, the magnitude of the coefficient, std error and the confidence intervals for X41 is exceptionally large as compared to the rest of the features. 

We could create a crosstable between X41 and Y to check if Y somehow separates X41.

```python
pd.crosstab(df['X41'], df['Y'])
```

```python
df[df['X41']>0.0]['Y'].value_counts()
```

We observe that for all values of X41 > 0.0, the corresponding value of Y is always 0. So clearly, basic logistic regression will not provide satisfactory results in this case. 

We have found at least one variable that's responsible for this separation of the dataset. But there might be many more. Since we have around 54 features, it's not viable to check the impact of each feature one by one. For this reason, we use the Firth Regression instead of the normal Logistic Regression. Firth regression is a form of penalized likelihood regression technique, and it works well in cases where there is separation. It basically employs a penalty term on the standard logistic function to generate parameter estimates and standard errors of a logistic regression model.

Since this model isn't available as such in Python, I shall be using John Lee's implementation of the same (link : https://gist.github.com/johnlees/3e06380965f367e4894ea20fbae2b90d) with some modifications to suit our needs.

# Benchmark Model using Firth Regression

```python
import FirthRegression as flr
```

```python
# We now fit the benchmark model using this technique
(intercept, beta, bse, fitll, summary) = flr.fit_firth(Y_train, X_train )
```

Comparing the stats with the benchmark model built using Logistic regression, we see that the coefficients have shrunk, and the standard errors are also not as high as they were, earlier. But there are features with p-value > 0.05, which suggests that they might not be significant, and hence we drop them. We create a list of features with p-value >0.05, drop them from the train set, fit the model again, and check the output.

```python
l_1 = summary[summary['Wald_p']>0.05]['Variables'].tolist()
len(l_1)
```

We observe that there are 27 features with p-value > 0.05. We drop these features and check the output.

```python
(intercept, beta, bse, fitll, summary) = flr.fit_firth(Y_train, X_train.drop(l_1, 1))
```

We still have some features that have a p-value > 0.05, hence we fit the model again after dropping these features.

```python
l_2 = summary[summary['Wald_p']>0.05]['Variables'].tolist()
len(l_2)
```

We drop this feature as well.

```python
(intercept, beta, bse, fitll, summary) = flr.fit_firth(Y_train, X_train.drop(l_1+l_2, 1))
```

We now observe that the p-values of all the features are < 0.05. Hence we stop our feature selection here. We create a final list of features that we select, and save the train set with the list of selected features as a modified train set.

We then test the performance of this model on the validation set with the selected list of features and check various metrics using the confusion matrix. The Script used to test the model performance is saved as **model_performance.ipynb**.


```python
features = X.drop(coll_cols+l_1+l_2,1).columns.tolist()
features
```

```python
# We save the train set with the selected features 
modified_train = X_train[features].copy()
modified_train['Y_train'] = Y_train.copy()
modified_train.to_csv("modified_train_set.csv", index = False)
```
